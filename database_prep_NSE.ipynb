```python
df = pd.DataFrame()
for name in tickerSymbols:
    df=df.append(pd.read_sql(f'SELECT Date,"Adj Close" as "{name}" FROM "{name}" WHERE Date>"2010-01-01"',engine))
```

```python
df=df.groupby("Date").sum()
```

```python
df.index=pd.to_datetime(df.index)
```

```python
mtlprices= df.resample('M').last()
```
